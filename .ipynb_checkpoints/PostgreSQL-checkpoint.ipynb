```python
import psycopg2
from psycopg2 import Error
import pandas as pd
import seaborn as snb
import matplotlib.pyplot as plt
import holidays
from datetime import date
feriados = holidays.Brazil()
import dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output
from jupyter_dash import JupyterDash
import plotly.graph_objects as go
from datetime import date
import datetime
```

```python
app = JupyterDash(__name__)
JupyterDash.infer_jupyter_proxy_config()
semana = df[(df['Abertura']>'2020-10-01')&(df['Abertura']<'2020-11-01')]
app.layout = html.Div(children=[html.H1('Relatórios GENC/HOB',style={'textAlign':'center','color':'#503D36','font-size': 40}),
                               html.Div([
                                   dcc.DatePickerRange(
                                   id='date-picker',
                                   display_format=('DD/MM/YYYY'), 
                                   min_date_allowed=date(2017,1,1),
                                   max_date_allowed=date.today()+datetime.timedelta(days=1),
                                   initial_visible_month=date.today(),
                                   start_date=date(date.today().year,date.today().month,1),
                                   end_date=date.today()                                       
                                   )
                               ],style={'textAlign':'center'}), 
                               html.Div([
                                   html.Div(dcc.Graph(id='tabela')),
                                   html.Div(dcc.Graph(id='fechamento_graf'))
                               ])])

@app.callback([Output(component_id='tabela',component_property='figure'),
              Output(component_id='fechamento_graf',component_property='figure')],
              [Input('date-picker','start_date'),
              Input('date-picker','end_date')])
# Define a função que irá gerar e retornar os gráficos exibidos
def get_graph(start_date,end_date):
    end_date = str(end_date)+' 23:59'
    dados = df[(df['Abertura']>=start_date)&(df['Abertura']<=end_date)] # Filtra a tabela total com os parâmetros passados pelo usuário
    tabela = go.Figure(data=go.Table(columnwidth = [5,15,10,15],
                                     header=dict(values=['OS','Setor','Equipamento','Abertura','Nome']),
                                     cells=dict(values=[dados.OS,dados.Setor,dados.Equipamento,dados.Abertura,dados.Responsável])))
    fechamento_graf=go.Figure(data=[go.Pie(labels=['Fechadas','Abertas'],values=dados['OS'].groupby(dados['Situação']).count())])
    return[tabela,fechamento_graf]
    
if __name__=='__main__':
    app.run_server(mode='inline',host='localhost')
```
